```python
import matplotlib.pyplot as plt
```

```python
x_axis = ['Jan', 'Feb', 'Mar', 'April', "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
#create the plot with the ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
plt.plot(x_axis, y_axis, marker='D', color="g", linewidth=2, label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
```

```python
plt.barh(x_axis, y_axis, color='m', label="Boston")
plt.gca().invert_yaxis()
plt.xlabel('Fare($)')
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
fig, ax =plt.subplots()
plt.barh(x_axis, y_axis, color='c', label="Chicago")
plt.gca().invert_yaxis()
plt.xlabel('Fare($)')
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color="r", label ="Chicago")
plt.xlabel("Fare($)")
plt.xlim(0,50)
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i *3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, 
           s=[i *5 for i in y_axis],
           facecolor = "skyblue",
           linewidth = 2,
           alpha = 0.8,
           edgecolor= 'k',
           label = "Boston"
          )
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    shadow=True,
    startangle=96,
    explode=explode_values,
    counterclock=False,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
